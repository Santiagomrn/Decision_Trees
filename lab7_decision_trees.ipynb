```
# Import pandas library 
import pandas as pd 
import math
# initialize list of lists 
data = [['yes','long','no', 'yes'], 
        ['no', 'short','yes', 'yes',], 
        ['yes', 'long','yes','no'],
        ['no','long','yes','yes'],
        ['yes','short','yes','no']] 
index=['OVERCOOKED_PASTA', 'WAITING_TIME','RUDE_WAITER','SATISFIED']

data2 = [['sunny','hot','high', 'weak','no'], 
        ['sunny','hot','high', 'strong','no'],
         ['overcast','hot','high', 'weak','yes'],
         ['rain','mild','high', 'weak','yes'],
         ['rain','cool','normal', 'weak','yes'],
         ['rain','cool','normal', 'strong','no'],
         ['overcast','cool','normal', 'weak','yes'],
         ['sunny','mild','high', 'weak','no'],
         ['sunny','cool','normal', 'weak','yes'],
         ['rain','mild','normal', 'weak','yes'],
         ['sunny','mild','normal', 'strong','yes'],
         ['overcast','mild','high', 'strong','yes'],
         ['overcast','hot','normal', 'weak','yes'],
         ['rain','mild','high','strong','no'],] 
index2=['OUTLOOK', 'TEMPERATURE','HUMIDITY','WIND','PLAY']

data3=[['green','3','mango'],
       ['yellow','3','mango'],
       ['red','1','grape'],
       ['red','1','grape'],
       ['yellow','3','lemon']]

index3=['COLOUR','DIAMETER','LABEL']
# Create the pandas DataFrame 
data = pd.DataFrame(data, columns =index) 
# print dataframe. 
data
```

```
class decision_tree:
    def __init__(self,data):
      self.data=data
    def find_on_tree(self,branch,features):
        #print(branch)
        if type(branch) is dict:
          for key in branch:
            #print(features[key])
            self.find_on_tree(branch[key][features[key]],features)
        else:
             print(self.column_target," :",branch)
             print(" ")   
    def predict(self,features):
        #print(self.tree)
        #print(features[self.root])
        print(features)
        self.find_on_tree(self.tree[self.root][features[self.root]],features)
    
    def get_best_feature(self,column_target,data):
      best_gain=0
      best_feature="none"
      #print(data)
      for feature in list(data):
        if feature != column_target:
          
          feature_gain=self.gain(feature,column_target,data)
          #print(feature," :",feature_gain)
          
        if best_gain < feature_gain:
          best_gain=feature_gain
          best_feature=feature
      if len(data)>=2 and best_feature=="none":
        print(list(data[column_target]))
      #print("best gain: ",best_gain, "best feature :",best_feature)
      #print(" ")
      return best_feature
    
    def create_tree(self,column_target,data,level):
      ident=""
      branch="-"
      for i in range(level*15):
        ident +=" "
      for i in range((level+1)*15-5):
        branch +="-"
      tree={}

      best_feature=self.get_best_feature(column_target,data)
      if best_feature=="none":
        return
      
      print(ident,best_feature)
      if level==0:
        self.root=best_feature
      tree[best_feature]={}
      for feature_value in data[best_feature].unique():
        
        if self.entropy_s(column_target,data[(data[best_feature]==feature_value)]) == 0:
          
          tree[best_feature][feature_value]=data[(data[best_feature]==feature_value)][column_target].unique()[0]
          #print the tree
          print(ident,"       ",feature_value,"-", tree[best_feature][feature_value])
        else:
          #print the tree
          print(ident,"       ",feature_value,branch)
          tree[best_feature][feature_value]=self.create_tree(column_target,data[(data[best_feature]==feature_value)].loc[:, data.columns != best_feature ],level+1)
        print(" ")
        #print(tree)
        #print(data[(data[best_feature]==feature_value)])
      return tree
      
    
    def train(self,column_target):
      self.column_target=column_target
      data=self.data
      self.tree=self.create_tree(column_target,data,0)
      #print(self.tree)
      

    def gain(self,feature,column_target,data):
      entropies=0;
      for feature_value in data[feature].unique():
        Sv=len(data[(data[feature]==feature_value)][feature])
        S=len(data[feature])
        entropy_f=self.entropy_feature(feature,feature_value,column_target,data)
        #print(entropy_f)
        entropies += Sv/S*entropy_f
      gain=self.entropy_s(column_target,data)-(entropies)
      return gain
    
    def entropy_s(self,column_target,data):
      values={}
      samples=0
      entropy=0
      for target in data[column_target].unique():
        values[target]=len(data[(data[column_target]==target)][column_target])
        samples +=values[target]
      #print(values,samples)
      for target in values:
        #print(values[target]/samples*math.log2(values[target]/samples))
        entropy +=values[target]/samples*math.log2(values[target]/samples)
      return -1*entropy

    def entropy_feature(self,feature,feature_value,column_target,data):
      values={}
      samples=0
      entropy=0
      for target in data[column_target].unique():
        #print(data[(data[column_target]==target) & (data[feature]==feature_value)][feature])
        values[target]=len(data[(data[column_target]==target) & (data[feature]==feature_value)][feature])
        samples +=values[target]
      #print(values,samples,feature_value)
      for target in values:
        if values[target]==0:
          entropy +=0
          #print(0)
        else:
          #print(values[target]/samples*math.log2(values[target]/samples))
          entropy +=values[target]/samples*math.log2(values[target]/samples)
      return -1*entropy

```

```
t=decision_tree(data)
t.train('SATISFIED')

#predict={'OUTLOOK':"sunny", 'TEMPERATURE':"hot",'HUMIDITY':"normal",'WIND':"weak"}
predict={'OVERCOOKED_PASTA':'no','WAITING_TIME':'short','RUDE_WAITER':'no'}
t.predict(predict)

predict={'OVERCOOKED_PASTA':'yes','WAITING_TIME':'long','RUDE_WAITER':'yes'}
t.predict(predict)

predict={'OVERCOOKED_PASTA':'yes','WAITING_TIME':'short','RUDE_WAITER':'no'}
t.predict(predict)
```
